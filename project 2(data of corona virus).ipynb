```python
response=requests.get("https://coronavirus.m.pipedream.net/")
```

```python
response.json()
```

```python
data=response.json()
data
```

```python
data["rawData"]
```

```python
row_data=data["rawData"]
row_data
```

```python
for e in row_data:
    print(e['Country_Region'],"|",e['Province_State'],"|",e['Confirmed'],"|",e['Deaths'])
```

```python
with open ("Corona Virus Data.csv","w",newline="") as f:
    writer= csv.DictWriter(f,fieldnames=["Country Region","Province State","Confirmed","Deaths"])
    writer.writeheader()
    for e in row_data:

        Country_Region_=e["Country_Region"]
        Province_State_=e["Province_State"]
        Confirmed_=e["Confirmed"]
        Deaths_=e["Deaths"]

        writer.writerow(
            {
              "Country Region": Country_Region_,
              "Province State":Province_State_,
              "Confirmed":Confirmed_,
              "Deaths":Deaths_
            }
        )
```
